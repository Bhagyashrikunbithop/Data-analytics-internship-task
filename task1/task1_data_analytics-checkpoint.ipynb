```python
# data profiting
print("\n--- Raw Data ---")
print("Missing values:")
print(df.isnull().sum())
print("Duplicate rows:")
print(df.duplicated().sum())
```

```python
# Removing duplicates
rows_before = len(df)
df.drop_duplicates(inplace=True)
rows_after = len(df)
print("\n--- After removing duplicates ---")
print("Rows removed:", rows_before - rows_after)
```

```python
# checking duplicates are removed 
print("Remaining duplicates:", df.duplicated().sum())
```

```python
# handling missing values from dataset
print("\n--- Missing values before handing ---")
print(df.isnull().sum())
```

```python
#filling missing bike perchased values with median
bike_col = 'past_3_years_bike_related_purchases'
if bike_col in df.columns :
    df[bike_col].fillna(df[bike_col].median())
```

```python
print("---- after handling missing values of bike perchase ----")
print('missing values are : ')
print(df[bike_col].isnull().sum())
```

```python
# drop rows where customer_id is missing
missing_before = df['customer_id'].isnull().sum()
df.dropna(subset = ['customer_id'], inplace = True)
missing_after = df['customer_id'].isnull().sum()
print("missing customer id before : ")
print(missing_before)
print("missing customer id after : ")
print(missing_after)
```

```python
# data standardization
print("----- Before data standardization of DOB -----")
print(df['DOB'])
df['DOB'] = pd.to_datetime(df['DOB'], errors = 'coerce')
print()
print("----- After data standardization of DOB -----")
print(df['DOB'])
```

```python
# text normalization 
text_columns = ['gender', 'job_title', 'job_industry_category', 'wealth_segment']

for col in text_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()

print("\n--- data normalizaion done ---")
```

```python
df
```

```python
print("---- feature engineering ----")
current_year = datetime.now().year
df['calculated_age'] = current_year - df['DOB'].dt.year
df['AgeGroup'] = pd.cut(
    df['calculated_age'],
    bins=[0, 18, 30, 45, 60, 100],
    labels=['Teen', 'Young Adult', 'Adult', 'Senior', 'Elder']
)

df['bike_perchase_category'] = pd.cut(df[bike_col], bins = [-1, 10, 30, 100],labels = ['low', 'medium', 'high'])
print("---- AgeGroup ----")
print(df['AgeGroup'] )
print("------------------------------------------")
print()
print("---- bike_perchase_category ----")
print(df['bike_perchase_category'])
```
